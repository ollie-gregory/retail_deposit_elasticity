```python
import pandas as pd
```

```python
df = pd.read_excel('data/raw_data.xlsx', sheet_name='Macroeconomic variables')
df = df.rename(columns={'Unnamed: 0': 'quarter'})
```

```python
df['quarter'] = pd.PeriodIndex(
    df['quarter'].str.replace(r'Q([1-4]) (\d{4})', r'\2Q\1', regex=True),
    freq='Q'
)
```

```python
df = df[['quarter', 'UK real GDP', 'UK CPI', 'UK unemployment rate', 'UK aggregate retail deposits', 'Bank Rate']]
df = df.set_index('quarter')
```

```python
df
```

```python
df['cpi_rate'] = ((df['UK CPI'] - df['UK CPI'].shift(4)) / df['UK CPI'].shift(4)) * 100
```

```python
df = df.rename(columns={
    'UK real GDP': 'real_GDP',
    'UK CPI': 'cpi_level',
    'UK unemployment rate': 'unemp_rate',
    'UK aggregate retail deposits': 'agg_deposits',
    'Bank Rate': 'base_rate'
})
```

```python
df.to_csv('data/data.csv')
```
